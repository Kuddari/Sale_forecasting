```python
#1. converting the wide data format to long data format and merged all the datasets
# ata sales, calendar, and sell prices to a single dataframe using common “d_” key (day number).
df_sales=pd.melt(sales, id_vars=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'], var_name='d', value_name='sold')
```

### 1. some info about  "wm_yr_wk" from prices dataframe:
1. wm_yr_wk represent week number.
2. in prices dataframe there is store_id,item_id,wm_yr_wk and sell_price. in that the sell_price mean that product average price on that perticular week.

### 2. strategy to assign price to each product. and explanation about this code line " pd.merge(df_final, prices, on=['store_id','item_id','wm_yr_wk'], how='left') "
3. so aim is to gather the all dataframe in single  dataframe right. so when it comes to prices dataframe we dont have explicit prise for each day insted we have price for each week.
4. for some product, price could be diferent for different state, store,period.
5. challenge is to assign right price to particular product considering state, store and  period.
**note: store id contain information about store id as well state id.**
6. so first we have to merge sales_calender dataframe with price based on state_is.
    note: store id contain information about store id as well state id. then  on store_id then item_id then finally on wm_yr_wk (week_id).  

```python
df_final= pd.merge(df_sales, calendar, on='d', how='left')
```

```python
df_final= pd.merge(df_final, prices, on=['store_id','item_id','wm_yr_wk'], how='left') 
```
